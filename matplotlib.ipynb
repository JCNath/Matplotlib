```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graph

```python
x=[0,1,2,3,4,5,6]
y=[0,1,4,9,16,25,36]

#Resize your graph
plt.figure(figsize=(5,3),dpi=150)
#plt.plot(x,y,label='random',color='yellow',linewidth=2,marker='.',linestyle='--',markersize=10,markeredgecolor='red')

#use shorthand notation
#color,marker,line
plt.plot(x,y,'y*--',label='x^2',markeredgecolor='r')

#Line #2
x2=np.arange(0,4,0.5)
plt.plot(x2[:6],x2[:6]**3,label='x^3')
plt.plot(x2[5:],x2[5:]**3,'--')
plt.title('First Graph!',fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4,5,6,7,8])
plt.yticks([0,1,4,9,16,25,36,49])

plt.legend()

plt.savefig('mygraph.png',dpi=300)

plt.show()
```

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(5,3),dpi=200)
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,3),dpi=200)
labels = ['A','B','C','D']
values = [2,4,8,6]
bars = plt.bar(labels,values,width=0.4,color='y',edgecolor='r')

bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('+')
bars[3].set_hatch('o')

plt.show()
```
