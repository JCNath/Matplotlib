```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Resize your graph
plt.figure(figsize=(5,3),dpi=150)

x=[0,1,2,3,4,5]
y=[0,2,4,6,8,10]
#plt.plot(x,y,color='red',linewidth='2',marker='*',markersize='12',label='2x',markeredgecolor='blue',linestyle='--')
plt.plot(x,y,'r*--',label='2x')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'b.-',label='x^^2')
plt.plot(x2[4:],x2[4:]**2,'b.--')

plt.title('Our First Graph!')

plt.xlabel('X axis !')

plt.ylabel('Y axis !')

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
labels = ['A','B','C','D']
values = [2,4,8,6]
bars = plt.bar(labels,values,width=0.4,color='grey',edgecolor='r')
bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('+')

bars[3].set_hatch('o')

plt.show()
plt.show()
```
